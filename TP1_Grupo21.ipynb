#### Ejercicio 2 - Repaso de Pandas
Realicen una estadistica descriptiva de ambas series de indicadores comparando los dos países.

```python
# Resolver acá
import pandas as pd
import wbdata

# Parámetros que buscamos

indicators = {"HD.HCI.AMRT.MA": "human_capital_index", "NY.GDP.PCAP.PP.KD": "gdppc"}
country = ["USA", "ARG"]
date = ("2010", "2020")
freq = 'Y'
parse_dates = True
keep_levels = False
skip_cache = False


df = wbdata.get_dataframe(indicators = indicators,
                     country=country,
                     date=date,
                     freq=freq,
                     parse_dates=parse_dates,
                     keep_levels=keep_levels,
                     skip_cache=skip_cache)

# Imprimimos los primeros datos
print(df)
```

```python
# Obtenemos estadísticas descriptivas
stats = df.describe().round(3)

# Mostramos resultados
print(stats)
```

#### Ejercicio 3 - Practicando con Matplotlib
Armen dos gráficos distintos usando la librería Matplotlib (repasen Clase 4). Uno programandolo con el estilo *pyplot* y otro gráfico de estilo *orientada a objetos*

```python
df_reset = df.reset_index()
```

```python
# Resolver acá estilo pyplot
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
```

```python
# Convertimos 'date' a formato datetime
df_reset["date"] = pd.to_datetime(df_reset["date"])

# Convertimos columnas numéricas, manejando valores erróneos
df_reset[["human_capital_index", "gdppc"]] = df_reset[["human_capital_index", "gdppc"]].apply(pd.to_numeric, errors='coerce')

# Interpolamos solo valores numéricos
df_reset[["human_capital_index", "gdppc"]] = df_reset[["human_capital_index", "gdppc"]].interpolate(method='linear')

print(df_reset.round(3))

```

```python
import matplotlib.pyplot as plt

# Configuración de los gráficos
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# --- 📈 Gráfico 1: GDP per capita ---
for country in df_reset["country"].unique():
    subset = df_reset[df_reset["country"] == country]
    axes[0].plot(subset["date"], subset["gdppc"], marker='o', linestyle="-", label=f"{country} - GDP per Capita")

axes[0].set_ylabel("GDP per Capita")
axes[0].set_title("Evolución del GDP per Capita")
axes[0].legend()
axes[0].grid()

# --- 📉 Gráfico 2: Human Capital Index ---
for country in df_reset["country"].unique():
    subset = df_reset[df_reset["country"] == country]
    axes[1].plot(subset["date"], subset["human_capital_index"], marker='s', linestyle="--", label=f"{country} - HCI")

axes[1].set_xlabel("Año")
axes[1].set_ylabel("Human Capital Index")
axes[1].set_title("Evolución del Human Capital Index")
axes[1].legend()
axes[1].grid()

# Mostramos los gráficos
plt.tight_layout()
plt.show()
```

```python
# Histogramas para ver distribución de valores
df.hist(figsize=(10, 4), bins=10, edgecolor='black')
plt.suptitle("Distribución de Human Capital Index y GDP per Capita")
plt.show()
```

```python
#  Diagrama de dispersión (scatter plot)
plt.figure(figsize=(6, 6))
sns.scatterplot(x=df_reset["human_capital_index"], y=df_reset["gdppc"], hue=df.index.get_level_values(0), palette="Set1")
plt.xlabel("Human Capital Index")
plt.ylabel("GDP per Capita")
plt.title("Relación entre HCI y GDP per Capita")
plt.grid()
plt.show()
```

Nota 1: es posible que logren obtener los links a las noticias sin el dominio: "https://www.lanacion.com.ar/". De ser así, concatenen el dominio a la ruta del link obtenido, tal que se obtenga un link al que se pueda acceder. Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
# Resolver acá

!pip install requests 
!pip install beautifulsoup4
!pip install pandas


import requests #html requestor
from bs4 import BeautifulSoup #html parser
import pandas as pd #dataframe manipulator

```

```python
#Le decimos de que pagina sacamos la información
url = "https://www.eldestapeweb.com/"

r = requests.get(url) #traigo el contenido del html
contenido = r.content

soup = BeautifulSoup(contenido, "html.parser")
soup

#Encontramos todos los enlaces
links = soup.find_all('a')


enlaces_sin_dominio = []
for link in links:
    href = link.get('href')
    if href and href.startswith('/'):
        enlaces_sin_dominio.append(href)

for enlace in enlaces_sin_dominio:
    print(enlace)

# Manejo de errores en caso de que la solicitud falle (lo sacamos de ChatGPT porque no nos dejaba por errores)
if r.status_code != 200:
    print(f"Error al acceder a la página: {r.status_code}")

```

```python
# Concatenar el dominio a las rutas de los enlaces
dominio = "https://www.eldestapeweb.com"
enlaces_con_dominio = [dominio + enlace for enlace in enlaces_sin_dominio]

# Imprimir los enlaces completos
for enlace in enlaces_con_dominio:
    print(enlace)

# Manejo de errores en caso de que la solicitud falle
if r.status_code != 200:
    print(f"Error al acceder a la página: {r.status_code}")
```

```python
# Creamos un DataFrame con los enlaces
df = pd.DataFrame(enlaces_con_dominio, columns=['Enlaces'])

# Exportamos a un archivo Excel
df.to_excel('enlaces_noticias.xlsx', index=False)

# Imprimimos el DataFrame (opcional)
print(df)

# Manejo de errores en caso de que la solicitud falle (tambien ChatGPT)
if r.status_code != 200:
    print(f"Error al acceder a la página: {r.status_code}")
```
